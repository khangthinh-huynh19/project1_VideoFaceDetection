```python
import os
path='./dataset/cropped_image/'

img_manager=[]
for entry in os.scandir(path):
    name=entry.path.split('/')[-1]
    img_manager.append(entry.path)
    count=1
    for img_path in os.scandir(entry.path):
        old_name=img_path.path
        new_name=img_path.path.split('\\')[-2] +'/'+ name +str(count) +'.png'
        print(new_name)
        os.rename(old_name,new_name)
        count+=1
```
